```python
from minion import objective_functions
from minion import optimisation
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

np.set_printoptions(precision=6)
%matplotlib inline
%matplotlib widget
```

```python
func = objective_functions.himmelblau
```

```python
x = np.linspace(-5, 5, 100, dtype=np.float64)
y = np.linspace(-5, 5, 100, dtype=np.float64)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis", alpha=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Multi-Dimensional Function")
```

```python
result = None
x_init = np.array([-1, -1], dtype=np.float64)
result, hess = optimisation.newton(func, x_init, check=False)
result.history
```

```python
x_init = np.array([-1, -1], dtype=np.float64)
result = optimisation.gradient_descent(func, x_init, gamma=0.1)
result.history
```

```python
x_init = np.array([-1, -1], dtype=np.float64)
result, h = optimisation.quasi_newton(func, x_init, check=False, beta=0.2, alpha=(-1, 1))
assert optimisation.is_pos_def(np.linalg.inv(h))
result.history
```

```python
x_init = np.array([-1, -1], dtype=np.float64)
result = optimisation.nelder_mead(func, x_init, alpha=1, gamma=2, rho=0.5, sigma=0.5, scale=1, tol=1e-12)
result.history
```
